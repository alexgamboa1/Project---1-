```python
# Reading the csv file 
filepath ="Event_Usa.csv"
eventUsa_df = pd.read_csv(filepath)
eventUsa_df.head()
```

```python
# Counting the events according to the category 
Total_event  = eventUsa_df.groupby(["EventCategory"])["EventName"].count()
Total_event
```

```python
# Converting the events into dataframe
Total_eventdf = Total_event.to_frame()
Total_eventdf.rename(columns={"EventName":"TotalCount"}, inplace =True)
Total_eventdf.reset_index(inplace = True)
Total_eventdf
```

```python
# Showing the bar graph having the events according to the count.  and saving the figure.
plt.bar(Total_eventdf["EventCategory"],Total_eventdf["TotalCount"])
plt.xlabel("Type of Events")
plt.ylabel("Total Count of Events")
plt.title("Type vs Total Count")
plt.savefig("Count of types of events")
```

```python
# Counting the events according to the genre 
genre_df =eventUsa_df.groupby(["Genre"])["EventName"].count()
genre_df
```

```python
# Converting genre_df into the dataframe and saving the fig
genre_info = genre_df.to_frame().reset_index()
genre_info
plt.figure(figsize=(20,6))
plt.bar(genre_info["Genre"], genre_info["EventName"])
plt.xlabel("Different Type Of genre")
plt.ylabel("Total Count")
plt.title("Genre Vs Total Count")
plt.savefig("Genre.png")
```

```python
# Getting the events detail according to the subgenre"
sub_genre = eventUsa_df.groupby(["SubGenre"])["EventName"].count()
sub_genre = sub_genre.to_frame().reset_index()
sub_genre
```

```python
# Getting the bar graph according to the subgenre
plt.figure(figsize =(20,6))
plt.bar(sub_genre["SubGenre"], sub_genre["EventName"])

plt.xlabel("Sub Genre Information")
plt.ylabel("Total Counts")
plt.title("Sub Genre vs Total Count")
plt.savefig("subgenre.png")
```

```python
# Creating the graph according to the timezone. It displays that NewYork time zone has more number of events as comparison to others. 
timeZone = eventUsa_df.groupby(["StartTimezone"])["EventName"].count()
timeZone = timeZone.to_frame().reset_index()
plt.figure(figsize = (20,6))
plt.bar(timeZone["StartTimezone"],timeZone["EventName"])
plt.xlabel("States")
plt.ylabel("Total Count")
plt.title("States vs Total Count")
plt.savefig("TimeZone.png")
```

```python
# Creating the graph according to the promoters. It display Broadway in hollywood  is in the top of promoting events, while Aeg live is the least in this category.
promotor_info =  eventUsa_df.groupby(["PromoterName"])["EventName"].count()
promotor_info = promotor_info.to_frame().reset_index()
plt.figure(figsize =(30,20))
plt.bar(promotor_info["PromoterName"], promotor_info["EventName"])
plt.xlabel("Promotor Info")
plt.ylabel("Total Counts")
plt.title("Promotor vs Total Count")
plt.savefig("Promoter.png")
```

```python
# Getting the day and month number from start date of the events
eventUsa_df.StartLocalDate = pd.to_datetime(eventUsa_df.StartLocalDate)
eventUsa_df["Day"] = eventUsa_df.StartLocalDate.dt.day
eventUsa_df["Month_num"] =eventUsa_df.StartLocalDate.dt.month
eventUsa_df.head()
```

```python
#Getting the information of the events according to the dates and later saving the figure.
day_df =eventUsa_df.groupby(["Day"])["EventName"].count()
day_df=day_df.to_frame().reset_index()
plt.figure(figsize =(20,6))
plt.tight_layout()
plt.bar(day_df["Day"], day_df["EventName"])
plt.xlabel("Day")
plt.ylabel("Total Counts")
plt.title("Day vs Total Count")
plt.savefig("Date.png")
```

```python
#Getting the event information according to the months.November has most number of events while september has least.
month_df =eventUsa_df.groupby(["Month_num"])["EventName"].count()
month_df=month_df.to_frame().reset_index()
plt.tight_layout()
plt.bar(month_df["Month_num"], month_df["EventName"])
plt.xlabel("Month")
plt.ylabel("Total Counts")
plt.title("month vs Total Count")
plt.savefig("Months.png")
```
